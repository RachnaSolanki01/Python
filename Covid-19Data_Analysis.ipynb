# COVID-19 Data Analysis and Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("CONVENIENT_global_confirmed_cases.csv")

df.head()
```

```python
print("rows, columns")
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df = df.iloc[1:]

df.head()
```

```python
#Renaming the column of country/region to date.
df.rename(columns={'Country/Region' : 'Date'}, inplace = True)
```

```python
#changing the column to dataformat
df['Date'] = pd.to_datetime(df['Date'], format= '%m%d%y')
```

```python
df.info()
```

```python
df.head()
```

## Latest COVID Cases for Every Country

```python
#Convert All Country Columns to Numbers

for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col],errors='coerce')
```

```python
#Latest Data

latest_data = df.iloc[-1]

latest_cases = latest_data.drop('Date')

df.head(10)
```

```python
#Checking type of cases if they are object or float
print(type(latest_cases))

latest_cases.head()
```

```python
#Converting from Datatype numeric digits of the cases
latest_cases = pd.to_numeric(latest_cases, errors='coerce')

latest_cases.dtype
```

```python

top10 = latest_cases.sort_values(ascending=False).head(10)

print(top10)
```

## Bar Chart

```python
plt.figure(figsize=(12,6))

top10.plot(kind='bar')


plt.title('Top 10 Countries by COVID Cases')
plt.xlabel('Countries')
plt.ylabel('Cases')

plt.xticks(rotation=45)

plt.show()
```

##### The United States reported the highest number of COVID-19 cases among the selected countries on the latest available date. The United Kingdom ranked second, followed by Russia and Belgium.

## Worldwide COVID Trend Over Time

```python
numeric_df = df.iloc[1:,1:].apply(pd.to_numeric, errors='coerce')

world_cases = numeric_df.sum(axis=1)

world_cases.head()
```

## Trend Graph

```python
plt.figure(figsize=(15,6))

plt.plot(df['Date'][1:], world_cases)

plt.title("Worldwide COVID Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Cases")

plt.grid(True)

plt.show()
```

##### Global COVID-19 cases increased rapidly after the initial outbreak. Multiple waves of infection were observed over time, indicating recurring surges across different regions.

## Pie Chart of Top 10 Countries

```python
plt.figure(figsize=(8,8))

top10.plot(kind='pie', autopct='%1.1f%%')

plt.title('Top 10 Countries Share of COVID Cases')
plt.ylabel('')

plt.show()
```

##### The United States contributed the largest share of COVID-19 cases among the top 10 countries, accounting for approximately 31.9% of reported cases. The United Kingdom was the second-highest contributor with 19.6%.
