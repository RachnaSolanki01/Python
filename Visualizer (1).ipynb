```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sales_dataset.csv")

print(df.head())
```

### Importing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("sales_dataset.csv")

print(df.head())
```

```python

class SalesDataAnalyzer:

    def __init__(self, data):

        self.data = data

        print("Sales Data Analyzer Initialized Successfully")


    def explore_data(self):

        print("\n===== Explore Data =====")

        print("1. First 5 Rows")
        print("2. Last 5 Rows")
        print("3. Column Names")
        print("4. Data Types")
        print("5. Dataset Info")
        print("6. Statistical Summary")

        choice = input("\nEnter your choice: ")

        if choice == "1":

            print(self.data.head())

        elif choice == "2":

            print(self.data.tail())

        elif choice == "3":

            print(self.data.columns)

        elif choice == "4":

            print(self.data.dtypes)

        elif choice == "5":

            print(self.data.info())

        elif choice == "6":

            print(self.data.describe())

        else:

            print("Invalid Choice")


    def clean_data(self):

        print("\n===== Handle Missing Data =====")

        print("1. Display Missing Values")
        print("2. Fill Missing Values with Mean")
        print("3. Drop Missing Values")

        choice = input("\nEnter your choice: ")

        if choice == "1":

            print(self.data.isnull().sum())

        elif choice == "2":

            self.data["Sales"] = self.data["Sales"].fillna(self.data["Sales"].mean())

            self.data["Profit"] = self.data["Profit"].fillna(self.data["Profit"].mean())

            print("\nMissing Values Filled Successfully")

        elif choice == "3":

            self.data = self.data.dropna()

            print("\nMissing Rows Dropped Successfully")

        else:

            print("Invalid Choice")


    def mathematical_operations(self):

        sales_array = np.array(self.data["Sales"])

        profit_array = np.array(self.data["Profit"])

        print("\n===== Mathematical Operations =====")

        print("\nSales + 100")
        print(sales_array + 100)

        print("\nSales * 2")
        print(sales_array * 2)

        print("\nSales + Profit")
        print(sales_array + profit_array)


    def search_sort_filter(self):

        print("\n===== Search, Sort and Filter =====")

        print("1. Search Product")
        print("2. Sort Sales Ascending")
        print("3. Sort Sales Descending")
        print("4. Filter High Sales")

        choice = input("\nEnter your choice: ")

        if choice == "1":

            product_name = input("Enter Product Name: ")

            print(self.data[self.data["Product"] == product_name])

        elif choice == "2":

            print(self.data.sort_values(by="Sales"))

        elif choice == "3":

            print(self.data.sort_values(by="Sales", ascending=False))

        elif choice == "4":

            print(self.data[self.data["Sales"] > 3000])

        else:

            print("Invalid Choice")


    def aggregate_functions(self):

        print("\nTotal Sales")
        print(self.data["Sales"].sum())

        print("\nAverage Sales")
        print(self.data["Sales"].mean())


    def statistical_analysis(self):

        print("\nStandard Deviation")
        print(self.data["Sales"].std())

        print("\nVariance")
        print(self.data["Sales"].var())


    def groupby_operations(self):

        print(self.data.groupby("Region")["Sales"].sum())


    def create_pivot_table(self):

        pivot = self.data.pivot_table(
            values="Sales",
            index="Region",
            aggfunc="sum"
        )

        print(pivot)


    def visualize_data(self):

        print("\n===== Data Visualization =====")

        print("1. Bar Plot")
        print("2. Line Plot")
        print("3. Scatter Plot")
        print("4. Histogram")

        choice = input("\nEnter your choice: ")

        if choice == "1":

            region_sales = self.data.groupby("Region")["Sales"].sum()

            plt.bar(region_sales.index, region_sales.values)

            plt.title("Total Sales by Region")

            plt.show()

        elif choice == "2":

            plt.plot(self.data["Sales"])

            plt.title("Sales Trend")

            plt.show()

        elif choice == "3":

            plt.scatter(self.data["Sales"], self.data["Profit"])

            plt.title("Sales vs Profit")

            plt.show()

        elif choice == "4":

            plt.hist(self.data["Sales"])

            plt.title("Sales Distribution")

            plt.show()

        else:

            print("Invalid Choice")


    def save_visualization(self):

        region_sales = self.data.groupby("Region")["Sales"].sum()

        plt.bar(region_sales.index, region_sales.values)

        plt.title("Saved Bar Plot")

        plt.savefig("bar_plot.png")

        plt.show()

        print("\nVisualization Saved Successfully")


    def __del__(self):

        print("Object Destroyed")
```

```python
analyzer = SalesDataAnalyzer(df)
```

```python

while True:

    print("\n========== Data Analysis & Visualization Program ==========")

    print("\nPlease select an option:")

    print("1. Explore Data")
    print("2. Handle Missing Data")
    print("3. Mathematical Operations")
    print("4. Search, Sort and Filter")
    print("5. Aggregation Functions")
    print("6. Statistical Analysis")
    print("7. GroupBy Operations")
    print("8. Pivot Table")
    print("9. Data Visualization")
    print("10. Save Visualization")
    print("11. Exit")

    choice = input("\nEnter your choice: ")

    if choice == "1":

        analyzer.explore_data()

    elif choice == "2":

        analyzer.clean_data()

    elif choice == "3":

        analyzer.mathematical_operations()

    elif choice == "4":

        analyzer.search_sort_filter()

    elif choice == "5":

        analyzer.aggregate_functions()

    elif choice == "6":

        analyzer.statistical_analysis()

    elif choice == "7":

        analyzer.groupby_operations()

    elif choice == "8":

        analyzer.create_pivot_table()

    elif choice == "9":

        analyzer.visualize_data()

    elif choice == "10":

        analyzer.save_visualization()

    elif choice == "11":

        print("\nExiting Program... Goodbye!")

        break

    else:

        print("\nInvalid Choice! Please Try Again.")
```
